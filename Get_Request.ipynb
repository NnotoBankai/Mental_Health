```python
import requests
```

```python
# The direct download link that I obtained
csv_url = 'https://data.cdc.gov/api/views/yni7-er2q/rows.csv?accessType=DOWNLOAD'
```

```python
# Send a GET request to the URL
response = requests.get(csv_url, allow_redirects=True)
```

```python
# Check if the request was successful
if response.status_code == 200:
    # Save the content of the request into a file
    with open('downloaded_dataset.csv', 'wb') as file:
        file.write(response.content)
    print("Dataset downloaded successfully.")
else:
    print("Failed to retrieve the dataset.")
```
